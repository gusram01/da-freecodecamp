```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    [
        {
            "product_id": 23,
            "name": "computer",
            "wholesale_price": 500,
            "retail_price": 1000,
            "sales": 100,
        },
        {
            "product_id": 96,
            "name": "Python Workout",
            "wholesale_price": 35,
            "retail_price": 75,
            "sales": 1000,
        },
        {
            "product_id": 97,
            "name": "Pandas Workout",
            "wholesale_price": 35,
            "retail_price": 75,
            "sales": 500,
        },
        {
            "product_id": 15,
            "name": "banana",
            "wholesale_price": 0.5,
            "retail_price": 1,
            "sales": 200,
        },
        {
            "product_id": 87,
            "name": "sandwich",
            "wholesale_price": 3,
            "retail_price": 5,
            "sales": 300,
        },
    ]
)
```

```python
net_income = (df["retail_price"] - df["wholesale_price"]) * df["sales"]

df_w_net = df.assign(
    current_net=net_income,
    after_15=net_income * 0.85,
    after_20=net_income * 0.8,
    after_25=net_income * 0.75,
)

df_w_net
```

```python
new_products = pd.DataFrame(
    [
        {
            "product_id": 24,
            "name": "phone",
            "wholesale_price": 200,
            "retail_price": 500,
        },
        {"product_id": 16, "name": "apple", "wholesale_price": 0.5, "retail_price": 1},
        {"product_id": 17, "name": "pear", "wholesale_price": 0.6, "retail_price": 1.2},
    ],
    index=range(5, 8),
)

df_w_new_prod = pd.concat([df, new_products])

df_w_new_prod
```

```python
df_w_new_prod.loc[[5,6,7], 'sales'] = [100, 200, 75]

df_w_new_prod
```
